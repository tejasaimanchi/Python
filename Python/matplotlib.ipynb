```python
import matplotlib.pyplot as plt
import pandas as pd
data={"days": ["A", "B", "C", "D"],"hours": [3, 5, 7, 9]}
df=pd.DataFrame(data)
ax = df.plot(kind='bar',x='days', y='hours', color='g')
ax.set_xlabel("Days")
ax.set_ylabel("Hours")
ax.set_title("Work")
plt.show()
```
